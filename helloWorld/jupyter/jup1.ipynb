```python
from matplotlib import pyplot as plt
```

```python
x=[1,2]
y=[3,4]
plt.plot(x,y)
```

```python
#x -100 - 100
x= [x for x in range(-100,101)]
y = [t**2 for t in x]
plt.plot(x,y)
```

```python
x= [x for x in range(-100,101)]
y = [t**3 for t in x]
plt.plot(x,y)
```

```python
x= [x+0.5 for x in range(-100,101)]
y = [1/t for t in x]
plt.plot(x,y)
```

```python
x= list(filter(lambda x: x!=0, [x+0.5 for x in range(-100,101)]))
y = [1/t for t in x]
plt.plot(x,y)
```

```python
import math
def f(x):
    return math.sin(x)
x= [x/100 for x in range(-314,314)]
y = [f(t) for t in x]
plt.plot(x,y)
```

```python
from sympy import *
from sympy.plotting import plot
```

```python
x = symbols('x')
p1 = plot(x*x - 5*x, show=false)
p2 = plot(5*x, show=false)
p1.append(p2[0])
p1.show()
```

```python
x = symbols('x')
p1 = plot(x*x, show = False)
p2 = plot(32*(-x)**0.5, show = False)
p3 = plot(-32*x**0.5, show = False)
p4 = plot(-32*(-x)**0.5, show = False)
p5 = plot(32*x**0.5, show = False)
p6 = plot(-(x*x), show = False)
p1.append(p2[0])
p1.append(p3[0])
p1.append(p4[0])
p1.append(p5[0])
p1.append(p6[0])
p1.show()
```
